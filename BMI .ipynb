```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/srishtichauhan/Desktop/BMIfinal.csv")
```

```python
df['BMI'] = ((df['Weight(kg)'])*10000)/((df['Height(cm)'])**2)
```

```python
df
```

```python
# 18.5 > undernourished
# 18.5 - 24.9 = healthy
# 25 - 29.9 = overweight
# 30 < obese
```

```python
df.describe()
```

```python
BMI = df['BMI'].values.tolist()
```

```python
cat = []
for i in range(0, len(BMI)):
    if (BMI[i] < 18.5):
        cat.append("Underweight")
    elif (18.5 <= BMI[i] <= 24.9):
        cat.append("Healthy")
    elif (25 <= BMI[i] <= 29.9):
        cat.append("Overweight")
    else:
        cat.append("Obese")
```

# GENDER BASED PLOTS 

```python
ds1 = df[df.Gender == "Female"]
ds2 = df[df.Gender == "Male"]
print(ds1.groupby(["Weight Category"]).count().plot(kind = "pie", y = "BMI", title = "FEMALE WEIGHT CATEGORY DISTRIBUTION")) # FEMALE
print(ds2.groupby(["Weight Category"]).count().plot(kind = "pie", y = "BMI", title = "MALE WEIGHT CATEGORY DISTRIBUTION")) # MALE
```
